```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/NLP/csvfiles/dataFile.csv')
```

```python
data = data.drop('Unnamed: 0',axis=1)
```

```python
y_train = data['label_b']
```

```python
data
```

```python
X_train = pd.DataFrame()
```

```python

X_train['ups'] = data['ups']
X_train['label_a'] = data['label_a']
X_train['text'] = data['text']
```

```python
X_train
```

```python
y_train = data['label_b'].astype(int)
```

```python
X_train = X_train.to_numpy()
X_train
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( X_train, y_train, test_size=0.20, random_state=42)
```

```python
x_train
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline, FeatureUnion

from sklearn.preprocessing import FunctionTransformer


# Create Function Transformer to use Feature Union
def get_numeric_data(x):
   # print([record[:-1].astype(float) for record in x])
    return [record[:-1].astype(float) for record in x]

def get_text_data(x):
    return [record[2] for record in x]

transfomer_numeric = FunctionTransformer(get_numeric_data)
transformer_text = FunctionTransformer(get_text_data)

# Create a pipeline to concatenate Tfidf Vector and Numeric data

pipeline = Pipeline([
    ('features', FeatureUnion([
            ('numeric_features', Pipeline([
                ('selector', transfomer_numeric)
            ])),
             ('text_features', Pipeline([
                ('selector', transformer_text),
                ('vec', TfidfVectorizer(analyzer='word'))
            ]))
         ])),
    ('clf',MLPClassifier(random_state=1, max_iter=300))
])



pipeline.fit(x_train, y_train)
```

```python
predictions = pipeline.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score -> ",accuracy_score(predictions, y_test)*100)
```

```python
from sklearn.metrics                  import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
class_probabilities = pipeline.predict_proba(x_test)
```

```python
class_probabilities
```
